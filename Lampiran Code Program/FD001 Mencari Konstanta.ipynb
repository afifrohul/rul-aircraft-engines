```python
import pandas as pd
from sklearn.feature_selection import VarianceThreshold
```

```python
df = pd.read_csv('CMAPSSData/train_FD001.csv')
df
```

```python
selector = VarianceThreshold(threshold=0.01)
X_selected = selector.fit_transform(df) 

selected_columns = df.columns[selector.get_support()]
print("Fitur yang dipertahankan:", selected_columns)
```

```python
removed_columns = df.columns[~selector.get_support()]
print("Fitur konstan yang dibuang:", removed_columns)
```
